## Nulls

```python
# Counting the nulls
world.isnull().sum(axis = 0)
```

```python
# There are too many columns for the previous method to show all nulls.

total_empty = [] # This will hold any cols completely empty

for col in world:
    nulls = world[col].isna().sum()
    if nulls == 0:
        pass
    elif nulls != len(world.index):
        print('Nulls in: ', col, ': ', nulls)
    elif nulls == len(world.index):
        total_empty.append(col) # This will place any cols completely empty in the 'empty' list.
        print('Nulls in: ', col, ': ', nulls, '   <----   This col is completely empty')
```

Given what we can see in the preview of the df, and that there between 5 and 3 rows in every column empty, we know it's the last few rows.

```python
world = world.drop([1064, 1065, 1066, 1067, 1068])
```

```python
world
```

```python
# '..' is a null? Counting to see how many in one column.
world[world['1960 [YR1960]'] == '..'].shape[0]
# Will change the '..' to a null which is in consistency with our democracy and co2 data.
```

```python
world = world.replace('..', np.nan)
```

```python
world
```

## Melt table

```python
# Before we melt the table it's a good idea to visualise what the table will look like.

datayears = range(1960, 2021) #our last col is for the year 2020.
no_years = len(datayears) # This is how many years we have data on
# And 1064 row
expected_rows = no_years*1064
expected_rows
```

```python
world_melted = pd.melt(world, 
            id_vars=["Country Name", "Country Code", "Series Name"],
            value_vars=list(world.columns[4:]),
            var_name='Year', 
            value_name='Value') ##seriesname value. Will pivot up again.
```

```python
world_melted
```

## Pivot table
Before we clean the year value, we need to move the series name into columns.

```python
world_pivoted = (world_melted.set_index(["Country Name", "Country Code", "Year"])
        .pivot(columns="Series Name")['Value']
        .reset_index()
        .rename_axis(None, axis=1)
     )

# Should include a test to see if this actually worked ok?
```

```python
world_pivoted
```

Clean year column up.

```python
world_pivoted['Year'] = world_pivoted['Year'].str[:-8]
world_pivoted
```

## Data Types

```python
world_pivoted.dtypes
```

'Year' should be a date, and thereafter should columns be numbers.

```python
world_pivoted['Year'] = pd.to_numeric(world_pivoted['Year'])
world_pivoted['Year'] = pd.to_datetime(world_pivoted['Year'], format='%Y')
world_pivoted['Compulsory education, duration (years)'] = pd.to_numeric(world_pivoted['Compulsory education, duration (years)'])
world_pivoted['GDP (current US$)'] = pd.to_numeric(world_pivoted['GDP (current US$)'])
world_pivoted['GDP per capita (current US$)'] = pd.to_numeric(world_pivoted['GDP per capita (current US$)'])
world_pivoted['GDP per capita growth (annual %)'] = pd.to_numeric(world_pivoted['GDP per capita growth (annual %)'])

world_pivoted
# Must check numeric values of GDP.. 
# Years should be an Int64 but because there are nulls we shall leave it as a Float64.
```
